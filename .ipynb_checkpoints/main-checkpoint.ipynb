## Generating Dataset Metadata

This step scans the FakeAVCeleb dataset directory and automatically generates a metadata file describing each video sample.

The script:
- Detects all category folders inside the dataset root directory.
- Infers **face manipulation labels** and **audio manipulation labels** based on folder names.
- Creates a structured record for every video in the dataset.

For each video, the following information is stored:
- **video_id** – unique identifier for the video
- **category_folder** – the dataset category folder containing the sample
- **relative_path** – path used to locate the video inside the dataset
- **face_label** – whether the visual content is `real` or `fake`
- **audio_label** – whether the audio is `real` or `fake`

All metadata entries are saved into a CSV file:
./fakeavceleb_metadata.csv


This metadata file is later used to:
- organize dataset samples
- assign labels during preprocessing
- build balanced training datasets
- support frame extraction and model training pipelines

````python

import pandas as pd

OUT_CSV = "./fakeavceleb_metadata.csv"

if not os.path.exists(DATA_ROOT):
    raise SystemExit(f"DATA_ROOT not found: {DATA_ROOT}")

categories = sorted([d for d in os.listdir(DATA_ROOT) if os.path.isdir(os.path.join(DATA_ROOT, d)) and not d.startswith('.')])
print("Detected category folders:", categories)

rows = []
for cat in categories:
    cat_path = os.path.join(DATA_ROOT, cat)
    # determine labels from folder name (simple)
    cat_low = cat.lower()
    face_label = "fake" if "fakevideo" in cat_low.replace(" ", "") or "fakevideo" in cat_low else ("fake" if "fake" in cat_low and "video" in cat_low else "real")
    audio_label = "fake" if "fakeaudio" in cat_low.replace(" ", "") or "fakeaudio" in cat_low else ("fake" if "fake" in cat_low and "audio" in cat_low else "real")
    # fallback simpler:
    if "fakevideo" in cat_low.replace(" ", "") or cat_low.startswith("fakevideo") or "fakevideo" in cat_low:
        face_label = "fake"
    if "fakeaudio" in cat_low.replace(" ", "") or cat_low.startswith("fakeaudio") or "fakeaudio" in cat_low:
        audio_label = "fake"
    # iterate samples
    for sample in sorted(os.listdir(cat_path)):
        if sample.startswith("."):
            continue
        rows.append({
            "video_id": sample,
            "category_folder": cat,
            "relative_path": sample,
            "face_label": face_label,
            "audio_label": audio_label
        })

df = pd.DataFrame(rows)
os.makedirs(os.path.dirname(OUT_CSV), exist_ok=True)
df.to_csv(OUT_CSV, index=False)
print(f"Metadata written to: {OUT_CSV}  (samples: {len(df)})")
display(df.head(10))
````

The metadata is cleaned and transformed to create labels for model training.

Balancing prevents the model from being biased toward one class.

````python
import pandas as pd
import os

DATA_ROOT = "./FakeAVCeleb_v1.2/FakeAVCeleb_v1.2"
META_CSV = os.path.join(DATA_ROOT, "meta_data.csv")

df = pd.read_csv(META_CSV)

# Fix folder path
df['folder_path'] = df['Unnamed: 9'].apply(
    lambda x: x[len("FakeAVCeleb/"):] if x.startswith("FakeAVCeleb/") else x
)

# Create REAL / FAKE label
df["video_label"] = df["type"].apply(
    lambda x: "REAL" if "RealVideo" in x else "FAKE"
)

real_df = df[df.video_label == "REAL"]
fake_df = df[df.video_label == "FAKE"]

N = min(len(real_df), len(fake_df))

balanced_df = pd.concat([
    real_df.sample(N, random_state=42),
    fake_df.sample(N, random_state=42)
]).reset_index(drop=True)

print("Balanced dataset size:", len(balanced_df))
print(balanced_df.video_label.value_counts())
````

### 3. Frame Extraction from Videos
Each video is converted into a set of representative frames.

The extraction function:
- Reads the video using **OpenCV**
- Uniformly samples up to **200 frames per video**
- Skips frames that are too similar to previously saved frames
- Saves frames to disk for later processing

This ensures:
- consistent temporal sampling
- reduced redundancy
- manageable dataset size

````python
import cv2
import numpy as np
print(cv2.__version__, np.__version__)
````

````python
import cv2
import os
import numpy as np

MAX_FRAMES = 200
DUP_THRESHOLD = 5  # mean pixel difference threshold to consider frames duplicates

def extract_uniform_frames(video_path, out_dir):
    """
    Extracts approximately MAX_FRAMES uniform frames from video_path,
    removing near-duplicate frames.
    Returns list of saved frame paths.
    """
    os.makedirs(out_dir, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if total_frames == 0:
        cap.release()
        return []

    step = max(1, total_frames // MAX_FRAMES)

    saved_paths = []
    prev_frame = None
    frame_idx = 0
    saved_idx = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frame_idx % step == 0:
            # check near duplicate
            if prev_frame is not None:
                diff = cv2.absdiff(cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY),
                                   cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY))
                mean_diff = np.mean(diff)
                if mean_diff < DUP_THRESHOLD:
                    frame_idx += 1
                    continue

            fname = f"frame_{saved_idx:05d}.jpg"
            path = os.path.join(out_dir, fname)
            cv2.imwrite(path, frame)
            saved_paths.append(path)
            saved_idx += 1
            prev_frame = frame

        frame_idx += 1

        if saved_idx >= MAX_FRAMES:
            break

    cap.release()
    return saved_paths
````

### 4. Removing Near-Duplicate Frames
Even after uniform sampling, consecutive frames may still look nearly identical.

To address this:
- A **perceptual hash (pHash)** is computed for each frame
- Frames with very similar hashes are removed

This keeps only visually distinct frames and improves training efficiency.


````python
from PIL import Image
import imagehash
import os

HASH_THRESHOLD = 5

def remove_duplicate_frames(frame_paths):
    """
    Remove near-duplicate frames using perceptual hashing (phash).
    Keeps first frame, removes subsequent frames that are too similar.
    Returns list of kept frame paths.
    """
    if not frame_paths:
        return []

    kept = [frame_paths[0]]
    try:
        last_hash = imagehash.phash(Image.open(frame_paths[0]))
    except Exception:
        last_hash = None

    for f in frame_paths[1:]:
        try:
            current_hash = imagehash.phash(Image.open(f))
        except Exception:
            continue

        if last_hash is None or abs(last_hash - current_hash) > HASH_THRESHOLD:
            kept.append(f)
            last_hash = current_hash
        else:
            # remove duplicate frame
            try:
                os.remove(f)
            except Exception:
                pass

    return kept
````

### 5. Face Detection with MTCNN
Deepfake artifacts typically appear in facial regions.  
Therefore, the pipeline detects faces in every extracted frame using **MTCNN**.

For each frame:
- The face bounding box is detected
- Coordinates are saved to a `detections.csv` file
- The first detected face is saved as:
  - `face_crop.jpg` (entire face)
  - `mouth_roi.jpg` (mouth region)

These outputs allow later stages to:
- crop faces consistently
- focus on manipulation artifacts around the mouth

````python
from tqdm import tqdm
from PIL import Image
from facenet_pytorch import MTCNN
import os
import shutil
import traceback
import torch

OUT_ROOT = "./processed"
os.makedirs(OUT_ROOT, exist_ok=True)

# Setup MTCNN
device = 'cuda' if torch.cuda.is_available() else 'cpu'
mtcnn = MTCNN(keep_all=False, device=device)

errors = []

for idx, row in tqdm(balanced_df.iterrows(), total=len(balanced_df), desc="Samples"):

    try:
        folder = os.path.join(DATA_ROOT, row.folder_path)
        video_path = os.path.join(folder, row.path)

        if not os.path.exists(video_path):
            errors.append((row.source, "video_not_found"))
            continue

        vid = row.source
        label = row.video_label

        out_dir = os.path.join(OUT_ROOT, label, vid)
        frames_dir = os.path.join(out_dir, "frames")
        os.makedirs(frames_dir, exist_ok=True)

        # 1) Extract uniform frames
        frames = extract_uniform_frames(video_path, frames_dir)

        # 2) Remove near-duplicate frames
        frames = remove_duplicate_frames(frames)

        # 3) Run MTCNN and save detections.csv
        detections_csv = os.path.join(out_dir, "detections.csv")
        face_crop_path = os.path.join(out_dir, "face_crop.jpg")
        mouth_path = os.path.join(out_dir, "mouth_roi.jpg")

        with open(detections_csv, "w") as fh:
            fh.write("frame,x1,y1,x2,y2,prob\n")
            first_saved = False
            for fp in frames:
                try:
                    img = Image.open(fp).convert("RGB")
                    boxes, probs = mtcnn.detect(img)
                    if boxes is None or len(boxes) == 0:
                        fh.write(f"{os.path.basename(fp)},,,,0\n")
                        continue

                    x1, y1, x2, y2 = [int(v) for v in boxes[0]]
                    prob = float(probs[0]) if probs is not None else 1.0
                    fh.write(f"{os.path.basename(fp)},{x1},{y1},{x2},{y2},{prob}\n")

                    # save first face + mouth crop for reference
                    if not first_saved:
                        face = img.crop((x1, y1, x2, y2))
                        face.save(face_crop_path)

                        w, h = face.size
                        mx1, my1 = int(0.15 * w), int(0.55 * h)
                        mx2, my2 = int(0.85 * w), int(0.95 * h)
                        mouth = face.crop((mx1, my1, mx2, my2))
                        mouth.save(mouth_path)
                        first_saved = True

                except Exception:
                    fh.write(f"{os.path.basename(fp)},error,,,,\n")

    except Exception as e:
        errors.append((row.source, str(e)))
        traceback.print_exc()

print("Done. Samples processed:", len(balanced_df) - len(errors))
if errors:
    print("Errors (first 20):")
    for e in errors[:20]:
        print(" -", e)
else:
    print("No errors logged.")
````

## Per-Frame Face and Mouth Region Extraction + Model Training Setup

After extracting frames and detecting faces, this stage converts the processed videos into a **per-frame dataset** suitable for deepfake classification and prepares the data pipeline for model training.

### 1. Per-Frame Face and Mouth Cropping

Each video in the processed dataset contains:
- extracted frames
- face bounding box detections (`detections.csv`)

Using these detections, the pipeline performs the following steps:

1. **Load frame and detection data**
   - Each row in `detections.csv` corresponds to one frame and its detected face bounding box.

2. **Crop the face region**
   - Bounding box coordinates `(x1, y1, x2, y2)` are used to crop the face from the frame.
   - A small **padding (5%)** is added to ensure the entire face is captured.

3. **Extract the mouth region**
   - Deepfake artifacts frequently appear around the **lip synchronization area**.
   - The lower portion of the face crop is extracted as a **mouth region of interest (ROI)**.

4. **Save cropped images**
   - Face crops and mouth crops are stored separately.

Each video contributes **up to 100 frames** to keep the dataset balanced and manageable.

````python
import os, glob
from PIL import Image
import pandas as pd
from tqdm import tqdm

PROCESSED_ROOT = "./processed"
PERFRAME_ROOT  = "./processed_perframe"

os.makedirs(PERFRAME_ROOT, exist_ok=True)

MAX_PER_VIDEO = 100

samples = []

# gather videos
for cat in sorted(os.listdir(PROCESSED_ROOT)):
    cat_path = os.path.join(PROCESSED_ROOT, cat)
    if not os.path.isdir(cat_path):
        continue

    for vid in sorted(os.listdir(cat_path)):
        vid_path = os.path.join(cat_path, vid)

        frames_dir = os.path.join(vid_path, "frames")
        detections = os.path.join(vid_path, "detections.csv")

        if os.path.exists(frames_dir) and os.path.exists(detections):
            samples.append((cat, vid, frames_dir, detections))

print("Samples:", len(samples))

# process
for cat, vid, frames_dir, detections in tqdm(samples):

    out_face_dir = os.path.join(PERFRAME_ROOT, cat, vid, "faces")
    out_mouth_dir = os.path.join(PERFRAME_ROOT, cat, vid, "mouth")

    os.makedirs(out_face_dir, exist_ok=True)
    os.makedirs(out_mouth_dir, exist_ok=True)

    try:
        df = pd.read_csv(detections)
    except:
        continue

    saved = 0

    for _, row in df.iterrows():
        if saved >= MAX_PER_VIDEO:
            break

        frame_name = row.get("frame")
        if not isinstance(frame_name, str) or frame_name == "":
            continue

        frame_path = os.path.join(frames_dir, frame_name)
        if not os.path.exists(frame_path):
            continue

        x1, y1, x2, y2 = row.get("x1"), row.get("y1"), row.get("x2"), row.get("y2")

        if pd.isna(x1) or pd.isna(y1) or pd.isna(x2) or pd.isna(y2):
            continue

        try:
            img = Image.open(frame_path).convert("RGB")

            # clamp to image bounds
            x1 = max(0, int(x1))
            y1 = max(0, int(y1))
            x2 = min(img.width, int(x2))
            y2 = min(img.height, int(y2))

            if x2 <= x1 or y2 <= y1:
                continue

            # optionally add padding around face
            pad_w = int((x2 - x1) * 0.05)
            pad_h = int((y2 - y1) * 0.05)
            x1p = max(0, x1 - pad_w)
            y1p = max(0, y1 - pad_h)
            x2p = min(img.width, x2 + pad_w)
            y2p = min(img.height, y2 + pad_h)

            face = img.crop((x1p, y1p, x2p, y2p))

            # save face
            face_path = os.path.join(out_face_dir, f"frame_{saved:05d}.jpg")
            face.save(face_path)

            # mouth region (lower part of face)
            w, h = face.size
            mx1 = int(0.15 * w)
            mx2 = int(0.85 * w)
            my1 = int(0.55 * h)
            my2 = int(0.95 * h)
            mouth = face.crop((mx1, my1, mx2, my2))

            mouth_path = os.path.join(out_mouth_dir, f"frame_{saved:05d}.jpg")
            mouth.save(mouth_path)

            saved += 1

        except:
            continue

print("Face + mouth crops created at:", PERFRAME_ROOT)
````

### 2. Output Dataset Structure

The resulting dataset is organized as:

````python
import os, glob, random
from PIL import Image
import torch
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
import timm
from sklearn.model_selection import train_test_split

# ---------------- CONFIG ----------------
PERFRAME_ROOT = "./processed_perframe"
BATCH_SIZE = 16
IMG_SIZE = 224
MAX_FRAMES_PER_VIDEO = 100
SEED = 42
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# ---------------- BUILD PER-FRAME DATA ----------------
rows = []
for label in ["REAL", "FAKE"]:
    label_dir = os.path.join(PERFRAME_ROOT, label)
    if not os.path.isdir(label_dir):
        continue
    for vid in os.listdir(label_dir):
        vid_dir = os.path.join(label_dir, vid, "faces")  # face crops
        if not os.path.isdir(vid_dir):
            continue
        imgs = sorted(glob.glob(os.path.join(vid_dir, "frame_*.jpg")))
        for p in imgs[:MAX_FRAMES_PER_VIDEO]:  # limit per video
            rows.append((p, 0 if label=="REAL" else 1, vid))  # now include vid

# ---------------- TRAIN/VAL SPLIT BY VIDEO ----------------
video_map = {}
for p, label, vid in rows:
    video_map.setdefault(vid, []).append((p, label))

video_list = list(video_map.keys())
train_v, val_v = train_test_split(
    video_list, test_size=0.15, random_state=SEED
)

train_entries = []
val_entries = []
for vid in video_map:
    if vid in train_v:
        train_entries += video_map[vid]
    else:
        val_entries += video_map[vid]

print(f"Train frames: {len(train_entries)}, Val frames: {len(val_entries)}")
print(f"Train videos: {len(train_v)}, Val videos: {len(val_v)}")

# ---------------- DATASET ----------------
class FrameDataset(Dataset):
    def __init__(self, entries, transform=None):
        self.entries = entries
        self.transform = transform
    def __len__(self):
        return len(self.entries)
    def __getitem__(self, idx):
        p, label = self.entries[idx][:2]  # ignore vid
        img = Image.open(p).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, torch.tensor(label, dtype=torch.long)

# ---------------- TRANSFORMS ----------------
train_tf = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.RandomHorizontalFlip(),
    T.ColorJitter(0.15, 0.15, 0.15, 0.05),
    T.RandomResizedCrop(IMG_SIZE, scale=(0.8, 1.0)),
    T.ToTensor(),
    T.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225])
])
val_tf = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.ToTensor(),
    T.Normalize([0.485,0.456,0.406], [0.229,0.224,0.225])
])

# ---------------- DATALOADERS ----------------
train_ds = FrameDataset(train_entries, transform=train_tf)
val_ds   = FrameDataset(val_entries, transform=val_tf)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=0, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=0, pin_memory=True)

print("Setup complete. Ready to train on per-frame crops.")
````

### Training the Visual Deepfake Detection Model

This step trains a **visual deepfake detection model** using the face crops extracted from each video frame. Each frame is treated as an individual training sample and labeled as **REAL (0)** or **FAKE (1)**.

The model uses **EfficientNet-B0** with pretrained ImageNet weights and is fine-tuned to classify frames as real or fake. Frames are split into **training and validation sets by video** to prevent data leakage.

During training:

* Images are resized to **224×224**
* Data augmentation (flip, color jitter, random crop) is applied
* The model is trained using **AdamW optimizer** and **cross-entropy loss**

After training, the model predicts a **fake probability for each frame**.
To obtain a **video-level prediction**, the probabilities of all frames in a video are averaged.

The trained model and the per-video prediction scores are saved for evaluation.


````python
# ---------------- Train visual model on per-frame face crops ----------------
!pip install -q timm

import os, glob
import random
import pandas as pd
from PIL import Image
import torch
from torch import nn, optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
import timm
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, confusion_matrix, classification_report
import numpy as np

# ---------------- CONFIG ----------------
PERFRAME_ROOT = "./processed_perframe"   # where face crops were saved
MODEL_OUT    = "./models/visual_effnet_b0_perframe.pth"
OUT_SCORE_CSV = "./models/visual_scores_perframe.csv"

BATCH_SIZE = 32
IMG_SIZE = 224
LR = 3e-4
EPOCHS = 6
SEED = 42
MAX_FRAMES_PER_VIDEO = 100
DEVICE = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")
NUM_WORKERS = 0  # important for notebooks to avoid BrokenPipeError

print("Device:", DEVICE)

# ---------------- Gather per-frame samples ----------------
rows = []

for label in ["REAL", "FAKE"]:
    label_path = os.path.join(PERFRAME_ROOT, label)
    if not os.path.isdir(label_path):
        continue
    for vid in sorted(os.listdir(label_path)):
        vid_path = os.path.join(label_path, vid, "faces")
        if not os.path.exists(vid_path):
            continue
        imgs = sorted(glob.glob(os.path.join(vid_path, "frame_*.jpg")))
        for p in imgs:
            rows.append((p, 1 if label=="FAKE" else 0, label, vid))  # (path, label, category, video_id)

print(f"Total per-frame samples collected: {len(rows)}")

if len(rows) < 20:
    raise SystemExit("Too few frame crops found. Ensure per-frame crops were created.")

# ---------------- Train/val split by video ----------------
video_map = {}
for p, label, cat, vid in rows:
    video_map.setdefault((cat, vid), []).append((p, label))

video_list = list(video_map.keys())

# stratify by fake/real based on first frame label
video_labels = [video_map[(cat,vid)][0][1] for cat,vid in video_list]

train_v, val_v = train_test_split(
    video_list,
    test_size=0.15,
    random_state=SEED,
    stratify=video_labels
)

train_entries, val_entries = [], []
for cat, vid in video_map:
    arr = video_map[(cat, vid)]
    if (cat, vid) in train_v:
        train_entries += arr
    else:
        val_entries += arr

print(f"Train frames: {len(train_entries)}, Val frames: {len(val_entries)}")
print(f"Train videos: {len(train_v)}, Val videos: {len(val_v)}")
print("Setup complete. Ready to train on per-frame crops.")

# ---------------- Dataset ----------------
class FrameDataset(Dataset):
    def __init__(self, entries, transform=None):
        self.entries = entries
        self.transform = transform
    def __len__(self):
        return len(self.entries)
    def __getitem__(self, idx):
        p,label = self.entries[idx][0], self.entries[idx][1]
        img = Image.open(p).convert('RGB')
        if self.transform:
            img = self.transform(img)
        return img, torch.tensor(label, dtype=torch.long)

train_tf = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.RandomHorizontalFlip(),
    T.ColorJitter(0.15,0.15,0.15,0.05),
    T.RandomResizedCrop(IMG_SIZE, scale=(0.8,1.0)),
    T.ToTensor(),
    T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
])
val_tf = T.Compose([
    T.Resize((IMG_SIZE, IMG_SIZE)),
    T.ToTensor(),
    T.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
])

train_ds = FrameDataset(train_entries, transform=train_tf)
val_ds   = FrameDataset(val_entries, transform=val_tf)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

# ---------------- Model ----------------
model = timm.create_model('tf_efficientnet_b0', pretrained=True, num_classes=2)
model.to(DEVICE)
criterion = nn.CrossEntropyLoss()
optimizer = optim.AdamW(model.parameters(), lr=LR, weight_decay=1e-4)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=2, gamma=0.5)

# ---------------- Train Loop ----------------
best_val = 0.0

for epoch in range(1, EPOCHS+1):
    model.train()
    running_loss = 0.0
    n = 0

    print(f"\nEpoch {epoch}/{EPOCHS} — Training:")
    
    for batch_idx, (imgs, labels) in enumerate(train_loader, 1):
        imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
        optimizer.zero_grad()
        out = model(imgs)
        loss = criterion(out, labels)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * imgs.size(0)
        n += imgs.size(0)

        if batch_idx % 50 == 0 or batch_idx == len(train_loader):
            avg_loss = running_loss / n
            print(f"  Batch {batch_idx}/{len(train_loader)} — Avg Loss: {avg_loss:.4f}")

    scheduler.step()
    train_loss = running_loss / n if n > 0 else 0

    # ---------------- Validation ----------------
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0

    print("  Validating...")
    with torch.no_grad():
        for imgs, labels in val_loader:
            imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
            out = model(imgs)
            loss = criterion(out, labels)
            val_loss += loss.item() * imgs.size(0)
            preds = out.argmax(1)
            correct += (preds == labels).sum().item()
            total += imgs.size(0)

    val_acc = correct / total if total > 0 else 0
    val_loss = val_loss / total if total > 0 else 0

    print(f"Epoch {epoch} complete — Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}")

    if val_acc > best_val:
        best_val = val_acc
        os.makedirs(os.path.dirname(MODEL_OUT), exist_ok=True)
        torch.save(model.state_dict(), MODEL_OUT)
        print(f"Saved best model so far with val_acc={best_val:.4f} to {MODEL_OUT}")

print(f"\nTraining finished. Best val acc (frame-level): {best_val:.4f}")

# ---------------- Per-video aggregation ----------------
print("Computing per-video scores (mean of frame probs)...")
model.eval()
video_results = []

with torch.no_grad():
    for (cat,vid), frames in video_map.items():
        folder = os.path.join(PERFRAME_ROOT, cat, vid, "faces")
        imgs = sorted(glob.glob(os.path.join(folder, "frame_*.jpg")))
        if len(imgs)==0:
            continue
        probs = []
        for p in imgs:
            img = Image.open(p).convert('RGB')
            inp = val_tf(img).unsqueeze(0).to(DEVICE)
            out = model(inp)
            prob = torch.softmax(out, dim=1)[0,1].item()  # prob of FAKE
            probs.append(prob)
        mean_prob = float(np.mean(probs))
        true_label = 1 if cat=="FAKE" else 0
        video_results.append({
            "video_id": vid, "category": cat, "true_label": true_label,
            "prob_fake": mean_prob, "n_frames": len(probs)
        })

vdf = pd.DataFrame(video_results)
os.makedirs(os.path.dirname(OUT_SCORE_CSV), exist_ok=True)
vdf.to_csv(OUT_SCORE_CSV, index=False)
print("Saved per-video visual scores to:", OUT_SCORE_CSV)

# ---------------- Simple metrics ----------------
y_true = vdf['true_label'].tolist()
y_prob = vdf['prob_fake'].tolist()
y_pred = [1 if p>=0.5 else 0 for p in y_prob]

try:
    auc = roc_auc_score(y_true, y_prob)
except:
    auc = float('nan')

print("Per-video samples:", len(vdf))
print("ROC AUC:", auc)
print("Confusion matrix:\n", confusion_matrix(y_true, y_pred))
print(classification_report(y_true, y_pred, digits=4))
````

Checking identity overlap between REAL and FAKE samples


````python
# Modified audio extraction + unique-id manifest writer
import os, glob, subprocess, shlex, re, csv
from pathlib import Path
from collections import Counter

ROOT = "./FakeAVCeleb_v1.2/FakeAVCeleb_v1.2"   # adjust if needed
OUT_ROOT = "./processed_audio"
MANIFEST_CSV = "audio_unique_manifest.csv"
os.makedirs(OUT_ROOT, exist_ok=True)

# find all videos (recursively)
video_exts = ("*.mp4","*.mov","*.avi","*.mkv","*.MP4","*.MOV","*.AVI","*.MKV")
videos = []
for ext in video_exts:
    videos += glob.glob(os.path.join(ROOT, "**", ext), recursive=True)
videos = sorted(videos)
print("Total video files found:", len(videos))
if len(videos) == 0:
    raise SystemExit("No video files found under ROOT="+ROOT)

# infer label from path
def infer_audio_label_from_path(p):
    lp = p.lower()
    if "fakeaudio" in lp:
        return 1, "FAKE"
    if "realaudio" in lp:
        return 0, "REAL"
    # fallback if unsure
    if "fakevideo" in lp:
        return None, None
    return None, None

def extract_audio_ffmpeg(video_path, out_wav_path, sample_rate=16000):
    os.makedirs(os.path.dirname(out_wav_path), exist_ok=True)
    cmd = f"ffmpeg -y -i {shlex.quote(video_path)} -ac 1 -ar {sample_rate} -vn -f wav {shlex.quote(out_wav_path)}"
    proc = subprocess.run(cmd, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    return proc.returncode == 0, proc.stderr.decode("utf-8", errors="ignore")

# control re-extraction
FORCE = False   # set True to re-extract all files

counts = Counter()
failed = []
manifest_rows = []

for vp in videos:
    lab, labname = infer_audio_label_from_path(vp)
    if lab is None:
        # If cannot infer, skip (print a short example once)
        # You can inspect these manually if any appear
        # (we'll skip silently here)
        continue

    # Extract a stable video_id token:
    # Prefer idNNNNN pattern anywhere in path; else use parent folder name; else use file stem.
    mm = re.search(r"(id\d{3,6})", vp, flags=re.IGNORECASE)
    if mm:
        vid_token = mm.group(1)
    else:
        parent = Path(vp).parent.name
        stem = Path(vp).stem
        vid_token = parent if parent else stem

    # make unique id by prefixing label name
    unique_id = f"{labname}__{vid_token}"

    # out path kept in label subfolder, preserve vid_token as folder to make browsing easy
    out_wav = os.path.join(OUT_ROOT, labname, vid_token, "audio.wav")

    if os.path.exists(out_wav) and not FORCE:
        counts[labname] += 1
        manifest_rows.append((unique_id, lab, labname, out_wav))
        continue

    ok, err = extract_audio_ffmpeg(vp, out_wav)
    if not ok:
        failed.append((vp, err))
    else:
        counts[labname] += 1
        manifest_rows.append((unique_id, lab, labname, out_wav))

print("Extraction pass finished.")
print("Counts (audio files extracted or already present):")
print("  REAL:", counts["REAL"])
print("  FAKE:", counts["FAKE"])
print("Total processed_audio items:", counts["REAL"] + counts["FAKE"])
if failed:
    print("Failed extractions:", len(failed))
    for p,err in failed[:5]:
        print(" - failed:", p)
        print("   ffmpeg stderr excerpt:", "\n".join(err.splitlines()[-6:]))

# Write manifest CSV
with open(MANIFEST_CSV, "w", newline="") as f:
    w = csv.writer(f)
    w.writerow(["unique_id","label_int","label_name","path"])
    for r in manifest_rows:
        w.writerow(r)

print("Wrote manifest:", MANIFEST_CSV, "rows:", len(manifest_rows))

# Optional: show number of base ids that appear in both REAL and FAKE (identity overlap)
base_ids_by_label = {"REAL": set(), "FAKE": set()}
for _,lab_int,labname,path in manifest_rows:
    # unique base id token is after the "__"
    base = path.split(os.sep)[-2]   # folder used for vid_token
    base_ids_by_label[labname].add(base)

dups = base_ids_by_label["REAL"].intersection(base_ids_by_label["FAKE"])
print("Number of base ids (identity tokens) present in BOTH REAL and FAKE folders:", len(dups))
if len(dups) > 0:
    print("Example duplicated base ids (up to 10):", list(dups)[:10])

print("Done. You can now use 'audio_unique_manifest.csv' or the variables above to build datasets with unique ids (label prefixed).")
````

# Wav2Vec2 Segment-Level Embedding Extraction

## Objective

This cell extracts **Wav2Vec2 embeddings** from processed audio files listed in the manifest.

Instead of computing a single embedding per audio file, we:

- Split each audio file into fixed-length segments  
- Extract embeddings per segment  
- Save them individually as `.npy` files  
- Allow resumable execution (skip already processed segments)

Processing Pipeline

### Load Manifest

Reads:

```python
audio_unique_manifest.csv
```

Each row contains:

- `unique_id`
- `label_int`
- `path`

---

### Audio Loading

Audio is:

- Converted to mono  
- Resampled to 16kHz (if needed)  
- Converted to `float32` NumPy array  

Supports:

- `soundfile + librosa`
- Fallback to `torchaudio`

---

### Segment Generation (Deterministic)

Segments are:

- Uniformly distributed across the full audio duration  
- Deterministically computed  
- Zero-padded if audio is shorter than segment length  

If audio length ≤ 1 second:

```
Only one padded segment is created.
```

---

### Embedding Extraction

For each batch of segments:

1. Tokenize using `AutoProcessor`
2. Forward pass through Wav2Vec2 encoder
3. Extract `last_hidden_state`
4. Apply **mean pooling over time**
5. Save embedding as `.npy`

Gradients are disabled using:

```python
@torch.no_grad()
```

---

## Output Structure

Embeddings are stored as:

```
audio_embeddings

````python
# A: Extract Wav2Vec2 embeddings per-segment (resumable, low-RAM)
import os, math, glob, random, numpy as np, torch, time
from pathlib import Path
from tqdm.auto import tqdm

# CONFIG
MANIFEST = "audio_unique_manifest.csv"   # or built earlier
OUT_DIR = "audio_embeddings"             # output folder (will be created)
SAMPLE_RATE = 16000
SEG_LEN_SEC = 1.0                        # 1s segments (changeable)
SEG_SAMPLES = int(SEG_LEN_SEC * SAMPLE_RATE)
N_SEGMENTS_PER_VIDEO = 10                # how many segments to extract per video (train use)
BATCH_SEG = 8                            # how many segments to forward at once through Wav2Vec2
DEVICE = torch.device("mps") if torch.backends.mps.is_available() else torch.device("cpu")
MODEL_NAME = "facebook/wav2vec2-base-960h"  # or your local checkpoint

os.makedirs(OUT_DIR, exist_ok=True)
print("Embedding extraction - SEG_LEN_SEC:", SEG_LEN_SEC, "SEG_SAMPLES:", SEG_SAMPLES, "BATCH_SEG:", BATCH_SEG)
print("Device for embedding extraction:", DEVICE)

# Load manifest
import pandas as pd
if not os.path.exists(MANIFEST):
    raise SystemExit("Manifest not found: " + MANIFEST)
dfm = pd.read_csv(MANIFEST)
items = [(row.unique_id, int(row.label_int), row.path) for row in dfm.itertuples(index=False)]
print("Manifest rows:", len(items))

# Processor & base model (create if missing)
from transformers import AutoProcessor, AutoModel
try:
    processor
except NameError:
    processor = AutoProcessor.from_pretrained(MODEL_NAME)

# create or reuse base (model.base should be the encoder)
try:
    base = model.base
except Exception:
    # load base model only (no head) - AutoModel gives last_hidden_state
    base = AutoModel.from_pretrained(MODEL_NAME)
base = base.to(DEVICE)
base.eval()

# helper audio loader
def load_audio(path, target_sr=SAMPLE_RATE):
    try:
        import soundfile as sf, librosa
        wav, sr = sf.read(path, always_2d=False)
        if wav.ndim > 1:
            wav = np.mean(wav, axis=1)
        if sr != target_sr:
            wav = librosa.resample(wav.astype(np.float32), orig_sr=sr, target_sr=target_sr)
        return wav.astype(np.float32)
    except Exception:
        import torchaudio
        wav_t, sr = torchaudio.load(path)
        wav = wav_t.mean(dim=0).numpy().astype(np.float32)
        if sr != target_sr:
            resampler = torchaudio.transforms.Resample(orig_freq=sr, new_freq=target_sr)
            wav = resampler(torch.from_numpy(wav)).numpy().astype(np.float32)
        return wav

# function to compute embedding for a batch of numpy audio segments
@torch.no_grad()
def embed_batch(seg_batch, device=DEVICE):
    # seg_batch: list of numpy arrays (each shape [seg_samples])
    inputs = processor(seg_batch, sampling_rate=SAMPLE_RATE, return_tensors="pt", padding=True)
    input_values = inputs.input_values.to(device)           # (B, T)
    attention_mask = inputs.attention_mask.to(device) if "attention_mask" in inputs else None
    out = base(input_values, attention_mask=attention_mask, return_dict=True)
    # out.last_hidden_state shape (B, seq_len, hidden)
    last = out.last_hidden_state   # tensor
    # mean-pool over time dimension (seq_len)
    emb = last.mean(dim=1).cpu().numpy()   # (B, hidden)
    return emb

# Build list of segment start times per video deterministically
def compute_starts(wav_len, seg_samples, n_segments):
    if wav_len <= seg_samples:
        return [0]
    # sample uniformly across available window
    max_start = wav_len - seg_samples
    if n_segments == 1:
        return [0]
    # spread evenly
    starts = [int(round(i * max_start / float(n_segments-1))) for i in range(n_segments)]
    return starts

# Main loop (resumable: skip existing files)
total = 0
for uid, lab, path in tqdm(items, desc="videos"):
    out_sub = os.path.join(OUT_DIR, uid)
    os.makedirs(out_sub, exist_ok=True)
    # compute a list of segment files already saved to skip
    existing = set([Path(p).stem for p in glob.glob(os.path.join(out_sub, "*.npy"))])
    # load wav
    wav = load_audio(path)
    n = wav.shape[0]
    starts = compute_starts(n, SEG_SAMPLES, N_SEGMENTS_PER_VIDEO)
    # build segment numpy list and target filenames
    segs = []
    fnames = []
    for i, s in enumerate(starts):
        fname = f"{uid}__seg{i:03d}.npy"
        if fname.replace(".npy","") in existing:
            continue
        e = s + SEG_SAMPLES
        if e <= n:
            seg = wav[s:e]
        else:
            seg = np.concatenate([wav[s:], np.zeros(e-n, dtype=np.float32)])
        segs.append(seg)
        fnames.append(os.path.join(out_sub, fname))
    # process in BATCH_SEG chunks
    for i in range(0, len(segs), BATCH_SEG):
        batch = segs[i:i+BATCH_SEG]
        emb = embed_batch(batch)   # (B, hidden)
        for j, vec in enumerate(emb):
            np.save(fnames[i+j], vec)
            total += 1

print("Embedding extraction complete. Saved embeddings:", total)
print("Example output folder:", os.path.join(OUT_DIR, items[0][0]) if items else "n/a")
````

# Per-Video Feature Aggregation (Mean Pooling)

## Objective

This cell aggregates previously extracted **segment-level Wav2Vec2 embeddings** into a single **per-video feature vector**.

Since each video was divided into multiple 1-second segments, we now:

- Load all segment embeddings per video  
- Stack them  
- Apply **mean pooling across segments**  
- Produce one fixed-size feature per video  

This enables fast, video-level classification experiments.

---
## Aggregation Process

For each `unique_id` in the manifest:

1. Locate its embedding folder:
   ```
   audio_embeddings/<unique_id>/
   ```

2. Load all `.npy` segment embeddings:
   ```
   (n_segments, 768)
   ```

3. Stack them into a matrix:
   ```python
   mats = np.vstack(mats)
   ```

4. Apply **mean pooling across segments**:
   ```python
   feat = mats.mean(axis=0)
   ```

Result:

```
Per-video feature shape: (768,)
```

---

## Why Mean Pooling?

Mean pooling:

- Produces fixed-length representations  
- Is simple and stable  
- Reduces segment-level noise  
- Works well as a strong baseline 


````python
# B: Aggregate segment embeddings into per-video features (mean pooling)
import os, glob, numpy as np, pandas as pd
from sklearn.model_selection import train_test_split

EMB_DIR = "audio_embeddings"
MANIFEST = "audio_unique_manifest.csv"   # same as used before
OUT_FEATURES = "video_features.npz"      # will save X,y,ids,labels

dfm = pd.read_csv(MANIFEST)
items = [(row.unique_id, int(row.label_int)) for row in dfm.itertuples(index=False)]
print("Manifest items:", len(items))

X = []
y = []
ids = []
missing = 0
for uid, lab in items:
    folder = os.path.join(EMB_DIR, uid)
    if not os.path.isdir(folder):
        missing += 1
        continue
    arr = sorted(glob.glob(os.path.join(folder, "*.npy")))
    if len(arr) == 0:
        missing += 1
        continue
    mats = [np.load(p) for p in arr]
    mats = np.vstack(mats)   # (n_segments, hidden)
    feat = mats.mean(axis=0)  # per-video mean embedding
    X.append(feat)
    y.append(lab)
    ids.append(uid)

print("Videos with embeddings:", len(X), "missing:", missing)
X = np.vstack(X)
y = np.array(y)
print("Feature shape:", X.shape, "Labels shape:", y.shape)

# Save features for quick experiments
np.savez_compressed(OUT_FEATURES, X=X, y=y, ids=np.array(ids))
print("Saved features to", OUT_FEATURES)
````

#C2 — Train a Small PyTorch MLP on Segment Embeddings

## Objective
Train a lightweight MLP classifier on precomputed **segment-level Wav2Vec2 embeddings** for fast experiments.  
Training is done on segments but evaluation is aggregated **per-video** (average segment probabilities) to avoid segment-level leakage.

---


````python

import os, glob, numpy as np, torch, random
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn, torch.optim as optim
from sklearn.metrics import roc_auc_score

EMB_DIR = "audio_embeddings"
# build list of (segment_path, label, video_id)
seg_items = []
for uid_lab in os.listdir(EMB_DIR):
    # uid_lab is unique id like REAL__id00431
    ufolder = os.path.join(EMB_DIR, uid_lab)
    if not os.path.isdir(ufolder): continue
    # label from uid_lab prefix
    lab = 1 if uid_lab.startswith("FAKE__") else 0
    for p in glob.glob(os.path.join(ufolder, "*.npy")):
        seg_items.append((p, lab, uid_lab))

print("Total segment embeddings:", len(seg_items))

# small dataset that loads .npy on the fly
class SegEmbDataset(Dataset):
    def __init__(self, items):
        self.items = items
    def __len__(self): return len(self.items)
    def __getitem__(self, idx):
        p, lab, vid = self.items[idx]
        x = np.load(p).astype(np.float32)
        return torch.from_numpy(x), torch.tensor(int(lab), dtype=torch.long), vid

# split by video to avoid leakage (important): build video-level split
from collections import defaultdict
videos = defaultdict(list)
for p,lab,uid in seg_items:
    videos[uid].append((p,lab,uid))
video_list = list(videos.items())
vids, labs = zip(*[(v, items[0][1]) for v, items in video_list])
# stratify by label
from sklearn.model_selection import train_test_split
vid_train, vid_val = train_test_split(list(video_list), test_size=0.15, random_state=42, stratify=[items[0][1] for _,items in video_list])
# flatten back to segment lists
train_segs = [it for v,its in vid_train for it in its]
val_segs = [it for v,its in vid_val for it in its]
print("Train segments:", len(train_segs), "Val segments:", len(val_segs))

train_ds = SegEmbDataset(train_segs)
val_ds = SegEmbDataset(val_segs)
train_loader = DataLoader(train_ds, batch_size=64, shuffle=True, num_workers=0)
val_loader = DataLoader(val_ds, batch_size=64, shuffle=False, num_workers=0)

# small MLP
hidden = train_ds[0][0].shape[0]
class MLP(nn.Module):
    def __init__(self, inp):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(inp, 512), nn.ReLU(), nn.Dropout(0.2),
            nn.Linear(512, 128), nn.ReLU(), nn.Dropout(0.1),
            nn.Linear(128, 2)
        )
    def forward(self, x):
        return self.net(x)

device = torch.device("cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu")
net = MLP(hidden).to(device)
opt = optim.AdamW(net.parameters(), lr=1e-3)
loss_fn = nn.CrossEntropyLoss()

# training loop (few epochs)
for epoch in range(1,6):
    net.train()
    total_loss = 0.0
    for xb, yb, _ in train_loader:
        xb = xb.to(device); yb = yb.to(device)
        opt.zero_grad()
        out = net(xb)
        loss = loss_fn(out, yb)
        loss.backward(); opt.step()
        total_loss += loss.item() * xb.size(0)
    print(f"Epoch {epoch} train loss:", total_loss / len(train_loader.dataset))

    # val: aggregate per-video by averaging probs across segments
    net.eval()
    import collections
    per_video = collections.defaultdict(list)
    with torch.no_grad():
        for xb, yb, vids in val_loader:
            xb = xb.to(device)
            logits = net(xb)
            probs = torch.softmax(logits, dim=1)[:,1].cpu().numpy()
            for p, v, lab in zip(probs, vids, yb):
                per_video[v].append((p, int(lab.item())))
    # aggregate
    y_true=[]; y_prob=[]
    for v, entries in per_video.items():
        labs = [e[1] for e in entries]
        # take majority label (should be consistent)
        y_true.append(labs[0])
        y_prob.append(np.mean([e[0] for e in entries]))
    try:
        auc = roc_auc_score(y_true, y_prob)
    except:
        auc = float("nan")
    print(" Val AUC (video):", auc)
````

````python
# SAVE AUDIO PER-VIDEO SCORES

import os
import numpy as np
import pandas as pd
import torch
from sklearn.metrics import roc_auc_score, confusion_matrix, classification_report

EMB_DIR = "audio_embeddings"
OUT_CSV = "audio_scores_pervideo.csv"

net.eval()

video_results = []

# Loop over each video folder in embeddings
for uid in os.listdir(EMB_DIR):
    folder = os.path.join(EMB_DIR, uid)
    if not os.path.isdir(folder):
        continue

    # label from uid prefix
    label = 1 if uid.startswith("FAKE__") else 0

    # load all segment embeddings
    files = sorted([os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(".npy")])
    if len(files) == 0:
        continue

    embs = np.vstack([np.load(f) for f in files]).astype(np.float32)

    with torch.no_grad():
        xb = torch.from_numpy(embs).to(device)
        logits = net(xb)
        probs = torch.softmax(logits, dim=1)[:,1].cpu().numpy()

    mean_prob = float(np.mean(probs))

    video_results.append({
        "video_id": uid,
        "true_label": label,
        "prob_fake": mean_prob,
        "n_segments": len(files)
    })

# Save CSV
vdf = pd.DataFrame(video_results)
vdf.to_csv(OUT_CSV, index=False)

print("Saved audio scores to:", OUT_CSV)
print("Total videos:", len(vdf))

# ---------- Quick Metrics ----------
y_true = vdf["true_label"].values
y_prob = vdf["prob_fake"].values
y_pred = (y_prob >= 0.5).astype(int)

try:
    auc = roc_auc_score(y_true, y_prob)
except:
    auc = float("nan")

print("\nPer-video ROC AUC:", auc)
print("Confusion matrix:\n", confusion_matrix(y_true, y_pred))
print(classification_report(y_true, y_pred, digits=4))
````

## Final Audio Model Results

This cell evaluates the final per-video predictions from the audio model.

It:
- Loads `audio_scores_pervideo.csv`
- Computes Accuracy and ROC-AUC
- Displays the Confusion Matrix
- Prints Precision, Recall, and F1-score

This provides the final performance summary of the audio deepfake detector before multimodal fusion.

````python
# Plot ROC curve for audio (and optionally visual / ensemble)
import os, numpy as np, pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, roc_auc_score

# --- Load audio per-video dataframe (use existing vdf if present) ---
try:
    vdf  # if vdf exists in kernel from earlier evaluation
except NameError:
    if os.path.exists("audio_scores_pervideo.csv"):
        vdf = pd.read_csv("audio_scores_pervideo.csv")
    else:
        raise SystemExit("No 'vdf' in kernel and no audio_scores_pervideo.csv file found. Run the save-scores cell first.")

# Expect vdf columns: video_id, true_label (or label), prob_fake (or prob)
# Normalize column names
if "true_label" in vdf.columns:
    y_col = "true_label"
elif "label" in vdf.columns:
    y_col = "label"
else:
    # try common fallbacks
    if "true" in vdf.columns[0].lower():
        y_col = vdf.columns[0]
    else:
        raise SystemExit("Can't find true label column in vdf. Expected 'true_label' or 'label'.")

if "prob_fake" in vdf.columns:
    p_col = "prob_fake"
elif "prob" in vdf.columns:
    p_col = "prob"
else:
    # try to locate a probability-like column
    cand = [c for c in vdf.columns if "prob" in c or "score" in c]
    if len(cand) > 0:
        p_col = cand[0]
    else:
        raise SystemExit("Can't find a probability column in vdf. Expected 'prob_fake' or 'prob'.")

y_true = vdf[y_col].values
y_score = vdf[p_col].values

# Compute ROC for audio
fpr_a, tpr_a, _ = roc_curve(y_true, y_score)
auc_a = auc(fpr_a, tpr_a)

# Try to load visual scores (optional) and compute ensemble (average) if possible
vis_loaded = False
if 'vdf_vis' in globals():
    vdf_vis = globals()['vdf_vis']
    vis_loaded = True
elif os.path.exists("visual_scores_pervideo.csv"):
    try:
        vdf_vis = pd.read_csv("visual_scores_pervideo.csv")
        vis_loaded = True
    except Exception:
        vis_loaded = False

plot_items = [("Audio", fpr_a, tpr_a, auc_a)]

if vis_loaded:
    # try to match on base id; allow video_id formats like REAL__idNNN or idNNN
    # normalize visual df
    vis = vdf_vis.copy()
    if "video_id" not in vis.columns:
        vis = vis.rename(columns={vis.columns[0]:"video_id"})
    # find visual prob column
    vis_prob_col = None
    for c in ("prob_fake","prob","prob_visual","score"):
        if c in vis.columns:
            vis_prob_col = c; break
    if vis_prob_col is None:
        # choose last column as fallback
        vis_prob_col = vis.columns[-1]
    # create base id keys
    def base_id_from_vid(vid):
        s = str(vid)
        if "__" in s:
            return s.split("__",1)[1]
        return s
    audio_df = vdf.copy(); audio_df["base_id"] = audio_df["video_id"].apply(base_id_from_vid)
    vis["base_id"] = vis["video_id"].apply(base_id_from_vid)
    merged = pd.merge(audio_df, vis[["base_id", vis_prob_col]].drop_duplicates(), on="base_id", how="inner")
    if len(merged) > 0:
        y_true_m = merged[y_col].values
        vis_scores = merged[vis_prob_col].values.astype(float)
        fpr_v, tpr_v, _ = roc_curve(y_true_m, vis_scores)
        auc_v = auc(fpr_v, tpr_v)
        plot_items.append(("Visual", fpr_v, tpr_v, auc_v))
        # ensemble by average
        ensemble_scores = (merged[p_col].values.astype(float) + vis_scores) / 2.0
        fpr_e, tpr_e, _ = roc_curve(y_true_m, ensemble_scores)
        auc_e = auc(fpr_e, tpr_e)
        plot_items.append(("Ensemble (avg)", fpr_e, tpr_e, auc_e))
    else:
        print("Visual scores present but no matching base ids found when merging. Skipping visual/ensemble plotting.")

# --- Plotting ---
plt.figure(figsize=(6,6))
for name, fpr, tpr, a in plot_items:
    plt.plot(fpr, tpr, label=f"{name} (AUC={a:.4f})")
plt.plot([0,1],[0,1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
out_png = "roc_audio.png"
plt.savefig(out_png, bbox_inches="tight")
plt.show()

print("Saved ROC plot to", out_png)
````
